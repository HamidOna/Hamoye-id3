```python
data_in = pd.read_excel("sonderauswertung-sterbefaelle.xlsx", sheet_name = 'D_2016-2021_Monate_AG_Ins', header = 8, index_col = 0,engine='openpyxl')

data_in = data_in.drop(['Unnamed: 15', 'Unnamed: 16'], axis = 1).reset_index(drop=True)
data_in.columns = ['year', 'age'] + MONTHS

# drop total
data_in = data_in[data_in.age != 'Insgesamt'].reset_index(drop=True)

# clean age column
data_in['age'] = data_in['age'].str.replace(' u. mehr', '+')
# define upper bound of age range
data_in['age_to'] = data_in['age'].map(AGE_MAP)
```

```python
data_in
```

```python
data = data_in.melt(id_vars=['year', 'age', 'age_to'] , value_vars = MONTHS).dropna()
data.columns = ['year', 'age', 'age_to', 'month', 'deaths']

# drop January and February 2016 to get four full years
data = data[(data['year']>2016) | (data['month'].isin(MONTHS[2:]))].reset_index(drop=True)

# mark covid period (2020/Mar - 2021/Feb) and non-covid period (2016/Mar - 2020/Feb)
data['period'] = '2020/Mar - 2021/Feb'
data.loc[(data.year < 2020) | ((data.year == 2020) & (data.month.isin(['Jan', 'Feb']))), 'period'] = '2016/Mar - 2020/Feb'
```

# Death by age

```python
fifteen = data[data["age"] == "0-15"].groupby('year')['deaths'].sum()
thirty = data[data["age"] == "15-30"].groupby('year')['deaths'].sum()
thirty_five = data[data["age"] == "30-35"].groupby('year')['deaths'].sum()
forty = data[data["age"] == "35-40"].groupby('year')['deaths'].sum()
forty_five = data[data["age"] == "40-45"].groupby('year')['deaths'].sum()
fifty = data[data["age"] == "45-50"].groupby('year')['deaths'].sum()
fifty_five = data[data["age"] == "50-55"].groupby('year')['deaths'].sum()
sixty = data[data["age"] == "55-60"].groupby('year')['deaths'].sum()
sixty_five = data[data["age"] == "60-65"].groupby('year')['deaths'].sum()
seventy = data[data["age"] == "65-70"].groupby('year')['deaths'].sum()
seventy_five = data[data["age"] == "70-75"].groupby('year')['deaths'].sum()
eighty = data[data["age"] == "75-80"].groupby('year')['deaths'].sum()
eighty_five = data[data["age"] == "80-85"].groupby('year')['deaths'].sum()
ninety = data[data["age"] == "85-90"].groupby('year')['deaths'].sum()
ninety_five = data[data["age"] == "90-95"].groupby('year')['deaths'].sum()
above_95 = data[data["age"] == "95+"].groupby('year')['deaths'].sum()
```

```python
age = pd.DataFrame({
    "fifteen": fifteen,
    "thirty": thirty,
    "thirty_five": thirty_five,
    "forty": forty,
    "forty_five": forty_five,
    "fifty": fifty,
    "fifty_five": fifty_five,
    "sixty": sixty,
    "sixty_five": sixty_five,
    "seventy": seventy,
    "seventy_five": seventy_five,
    "eighty": eighty,
    "eighty_five": eighty_five,
    "ninety": ninety,
    "ninety_five": ninety_five,
    "above_95": above_95
})
```

```python
plt.figure(figsize=(2,3))
plt.style.use('seaborn-bright')
age.plot.bar()
plt.legend(loc=1, prop={'size': 9})
plt.title('Death rate across age groups', weight='bold', size=23)
plt.ylabel('Death count')
```

Generally, ages above 65 have high deat count yearly. In 2016, age bracket 85-90 has the highest death count. In subsequent years, we see the age bracket eighty to eight five slowly become the age range with the most death count. Year 2021 has a low death count because it only contains data from january to february of the year.

```python
age.plot.bar(stacked=True)
age_total = age["fifteen"] + age["thirty"] + age["thirty_five"] + age["forty"] + age["forty_five"] + age["fifty"] + age["fifty_five"] + age["sixty"] + age["sixty_five"] + age['seventy'] + age["seventy_five"] + age["eighty"] + age['eighty_five'] + age['ninety'] + age['ninety_five'] + age['above_95']
age_rel = age[age.columns[1:]].div(age_total, 0)*100
plt.title('Death rate across age groups (stacked)', weight='bold', size=23)
plt.ylabel('Death count')
for n in age_rel:
    for i, (cs, ab, pc) in enumerate(zip(age.iloc[:, 1:].cumsum(1)[n], 
                                         age[n], age_rel[n])):
        if pc > 3:
            plt.text(i, cs - ab / 2, str(np.round(pc, 1)) + '%', 
                 va = 'center', ha = 'center')
```

We create a stacked bar chart to visualize better the difference in death count across the years. The death count only relatively increased slightly between 2019 and 2020 even though the world was afflicted with the covid-19 pandemic in year 2020.
